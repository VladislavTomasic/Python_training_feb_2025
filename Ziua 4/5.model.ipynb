```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x=[]
y=[]
plt.scatter(x,y)
plt.xlabel("pret")
plt.ylabel("m2")
plt.show()
```

```python
apartamente=[
    (70,100), (70,150), (100,120), (30,50), (50,70), (50,80), (100,500)
]

x=[ pair[0] for pair in apartamente ]
y=[ pair[1] for pair in apartamente ]
x,y
```

```python
plt.scatter(x,y)
plt.xlabel("pret")
plt.ylabel("m2")
plt.show()
```

```python
def functie_liniara(x):
    return x*1.5

y_prezis = [functie_liniara(i) for i in x]

plt.plot(x,y_prezis)
plt.scatter(x,y)
plt.xlabel("pret")
plt.ylabel("m2")
plt.show()
```

```python
from scipy import stats
```

```python


rezultat=stats.linregress(x,y)
rezultat

slope, intersept=rezultat.slope, rezultat.intercept

def functie_liniara(x):
    return intersept + x*slope

y_prezis = [functie_liniara(i) for i in x]

plt.plot(x,y_prezis)
plt.scatter(x,y)
plt.xlabel("pret")
plt.ylabel("m2")
plt.show()
```
